> **Ejercicio**: Diseñe una función que reciba una lista de palabras y devuelva, simultáneamente, la primera y la última palabra según el orden alfabético.

```python
def palabras(listapalabras):
  
  #primero pasamos todo a minusculas ò tambien puede ser a mayusculas, el caso es que queden todas iguales

  listapalabrasminus=[]
  for palabra in listapalabras:
        listapalabrasminus.append(palabra.lower())
    
  lista_ordenada = sorted(listapalabrasminus) #Hay funciones que ordenan, sorted
  

  primerapalabra=lista_ordenada[0]
  segundapalabra=lista_ordenada[-1]
  return primerapalabra,segundapalabra

#prueba

datos=['amarillo','Azul','Magenta','Verde','rojo']
primera_ejemplo,ultima_ejemplo= palabras(datos)
print('La primera es', primera_ejemplo, 'y la ultima',ultima_ejemplo)




```
